```python
import pandas as pd
import numpy as np
from pandas_datareader import data
import pandas_datareader.data as web
import datetime
```

```python
start =datetime.datetime(2020,3,31)
end=datetime.datetime(2021,4,13)
```

```python
stock='3481'+'.tw'
```

```python
#讀取資料
data=web.DataReader(stock,'yahoo',start,end)
#將空值與Adj Close去除
data=data.dropna()
data=data.drop(labels=['Adj Close'],axis=1)
#Volume比實際大1000倍左右，將它還原
trash=(data['Volume']%1000)
Volume=(data['Volume']-trash)/1000
data['Volume']=Volume

High_Range=[0]
Low_Range=[0]
Open_Range=[0]
Close_Range=[0]


for i in range(1,len(data)):
    H=round(((data['High'][i]-data['Close'][i-1])/data['High'][i-1]*100),2)
    L=round(((data['Low'][i]-data['Close'][i-1])/data['Low'][i-1]*100),2)
    O=round(((data['Open'][i]-data['Close'][i-1])/data['Close'][i-1]*100),2)
    C=round(((data['Close'][i]-data['Close'][i-1])/data['Close'][i-1]*100),2)
    High_Range.append(H)
    Low_Range.append(L)
    Open_Range.append(O)
    Close_Range.append(C)
    

data['High_Range']=High_Range
data['Low_Range']=Low_Range
data['Open_Range']=Open_Range
data['Close_Range']=Close_Range

data=data.drop(data.index[0])


data.to_csv(stock+".csv", encoding="utf_8_sig")
```
